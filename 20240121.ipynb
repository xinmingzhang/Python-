````python
import numpy as np
import matplotlib.pyplot as plt

# 生成 x 值，通常在 -2π 到 2π 之间
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# 计算正弦和余弦值
sin_y = np.sin(x)
cos_y = np.cos(x)

# 创建一个图形对象
plt.figure(figsize=(8, 6))

# 绘制正弦曲线，指定标签为 "Sin(x)"
plt.plot(x, sin_y, label="Sin(x)", linestyle="-", linewidth=2)

# 绘制余弦曲线，指定标签为 "Cos(x)"
plt.plot(x, cos_y, label="Cos(x)", linestyle="--", linewidth=2)

# 添加图例
plt.legend()

# 设置图形标题和坐标轴标签
plt.title("Sine and Cosine Curves")
plt.xlabel("x")
plt.ylabel("y")

# 显示图形
plt.grid(True)
plt.show()
````

## python中常用的数据结构

在Python中，常见的数据结构包括列表（List）、元组（Tuple）、字典（Dictionary）和集合（Set）。以下是它们的对比：

| 数据结构   | 定义                                          | 常用语句                             | 特点                                 | 常用操作                                     |
| --------- | ------------------------------------------- | ---------------------------------- | ------------------------------------ | ---------------------------------------- |
| 列表（List）   | 有序的可变容器，用方括号表示                      | `my_list = [1, 2, 3]`             | 允许重复元素，可以进行增删改查操作             | `len()`, `append()`, `extend()`, `remove()`, `pop()`, 索引和切片等 |
| 元组（Tuple） | 有序的不可变容器，用圆括号表示                      | `my_tuple = (1, 2, 3)`            | 允许重复元素，不可修改元素，可以进行索引和切片操作     | `len()`, 索引和切片等                            |
| 字典（Dictionary） | 无序的键值对的容器，用花括号表示                      | `my_dict = {'name': 'Alice', 'age': 25}` | 键值对唯一，通过键进行引用和修改值               | `len()`, `keys()`, `values()`, `items()`, `get()`, `pop()`, 添加和删除键值对等 |
| 集合（Set）   | 无序的唯一元素的容器，用花括号或`set()`函数进行创建              | `my_set = {1, 2, 3}`               | 不允许重复元素，可以进行集合运算（交、并、差）等操作 | `len()`, `add()`, `remove()`, 集合运算等            |

这只是一份简单的对比表格，每种数据结构在实际使用中还有更多的特性和方法。


## python中的“.”，“函数”

在Python中，`.`操作符用于访问对象的属性和方法。这使得我们可以通过对象名称后跟`.`符号来访问其属性和方法。

**访问属性**

当我们有一个对象时，可以使用`.`操作符来访问对象的属性。属性是存储在对象中的变量，可以是内置类型（例如数字、字符串等）或自定义类型（例如类的实例或对象）。

```python
person = {
  "name": "Alice",
  "age": 30,
  "country": "USA"
}

print(person.name)    # 访问对象的属性
print(person.age)
```

在这个例子中，`person`是一个字典对象，我们使用`.`操作符访问其属性`name`和`age`。输出结果为`Alice`和`30`。

**调用方法**

除了访问属性，我们还可以使用`.`操作符来调用对象的方法。方法是与对象关联的函数，它们用于执行特定的操作或行为。

```python
class Rectangle:
  def __init__(self, width, height):
    self.width = width
    self.height = height

  def area(self):
    return self.width * self.height

rect = Rectangle(5, 3)
print(rect.area())   # 调用对象的方法
```

在这个例子中，我们定义了一个名为`Rectangle`的类，它表示一个矩形对象。该类有一个`area`方法，用于计算矩形的面积。我们创建了一个`rect`对象，并使用`.`操作符调用了它的`area`方法。输出结果为`15`。

**模块和包**

在Python中，`.`操作符还用于访问模块和包中的函数、类和变量。模块是一个包含Python代码的文件，而包是包含多个模块的文件夹。

```python
import math    # 导入math模块

print(math.pi)    # 访问模块中的变量
print(math.sqrt(16))   # 调用模块中的函数
```

在这个例子中，我们使用`.`操作符来访问`math`模块中的常量`pi`和函数`sqrt`。输出结果为`3.141592653589793`和`4.0`。

通过`.`操作符，我们可以方便地访问对象的属性和方法，调用函数和类，以及导入和使用模块。这使得Python非常灵活和易于使用。

在Python中多个`.`操作符按从左到右的顺序执行。这意味着左边的`.`操作先执行，然后是右边的`.`操作。让我们通过一个例子来说明多个`.`操作符的运算顺序：

```python
class A:
    def __init__(self, x):
        self.x = x

    def method(self):
        print("Method A")

class B:
    def __init__(self, a_obj):
        self.a_obj = a_obj

    def method(self):
        print("Method B")

a = A(10)
b = B(a)
```

在上面的例子中，我们定义了两个类 `A` 和 `B`。类 `A` 有一个属性 `x` 和一个方法 `method`，类 `B` 有一个属性 `a_obj` 和一个方法 `method`。我们创建了一个 `A` 类的对象 `a`，然后将它作为参数传递给 `B` 类的对象 `b` 的构造函数。

现在，我们可以通过多个`.`操作符来进行运算。让我们看看不同的情况下运算的顺序：

```python
a.x             # 访问对象 a 的属性 x

b.a_obj         # 访问对象 b 的属性 a_obj
b.a_obj.x       # 访问对象 b 的属性 a_obj 的属性 x

a.method()      # 调用对象 a 的方法 method
b.method()      # 调用对象 b 的方法 method
b.a_obj.method()    # 调用对象 b 的属性 a_obj 的方法 method
```

根据从左到右的运算顺序，上述代码分别访问了对象 `a` 的属性 `x`，对象 `b` 的属性 `a_obj`，对象 `b.a_obj` 的属性 `x`，并调用了对象 `a` 和 `b` 的方法 `method`。

**python中函数返回值**

在Python函数中，返回值是指函数返回给调用者的结果。通过使用关键字`return`，函数可以返回一个值或一个对象。当函数执行到`return`语句时，它将会结束执行并返回指定的值。

以下是一个例子来说明Python函数的返回值：

```python
def add_numbers(a, b):
    sum = a + b
    return sum

result = add_numbers(3, 5)
print(result)  # 输出 8
```

在上面的例子中，我们定义了一个名为`add_numbers`的函数，它接受两个参数 `a` 和 `b`。函数将这两个参数相加得到 `sum` 的结果，并使用`return`关键字返回了这个结果。在调用`add_numbers`函数时，我们将其返回值赋给变量`result`，然后使用`print`函数打印出`result`的值，即得到了`3 + 5`的和`8`。

需要注意的是，函数可以返回任何类型的值，包括数字、字符串、列表、字典、甚至是其他函数或对象。函数也可以没有返回值，如果没有明确使用`return`语句，则函数将默认返回`None`。例如：

```python
def print_message():
    message = "Hello, world!"
    print(message)

result = print_message()
print(result)  # 输出 None
```

在这个例子中，函数`print_message`并没有使用`return`语句来返回任何值。它只是在屏幕上打印了一条消息。因此，当我们将其返回值赋给变量`result`时，`result`将被赋值为`None`，并且打印`None`。

````python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
my_number = factorial(1)
print(my_number)
````
